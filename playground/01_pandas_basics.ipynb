## 3.loc/iloc/布尔索引

```python
print("---loc['two']---")
print(frame2.loc['two'])
```

```python
print("\n---iloc[1]---")
print(frame2.iloc[1])
```

```python
print(frame2.loc['two', ['year', 'pop']])
```

```python
print(frame2.iloc[1, [0, 2]])
```

```python
print(frame2.iloc[[0,2],1]) 
```

```python
print(frame2[frame2['pop'] > 2])
```

```python
print(frame2[(frame2['pop'] > 2) & (frame2['year'] > 2001)])
```

```python
frame2.loc[frame2['pop'] > 2, ['city', 'year']]
```
